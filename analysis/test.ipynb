```python
from gramformer import Gramformer
import pandas as pd
```

```python
gf = Gramformer(models = 1, use_gpu=False) # 1=corrector, 2=detector
```

```python
df = pd.read_excel("test.xlsx")
df['edits'] = ''
df.head(3)
```

```python
for i,row in df.iterrows():

    ori_text=row['ori_text']
    cor_text=row['cor_text']

    edits=gf.get_edits(ori_text,cor_text)
    df.at[i, 'edits'] = edits
```

```python
for i,row in df.iterrows():

    print(i)
    for item in row['edits']:
        print(item)

    print("++++++++++++++++++++++++++++++++++++++++++++++++++")
```
